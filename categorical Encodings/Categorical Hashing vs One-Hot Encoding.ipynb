#### Feature Hashing Python

```python
import pandas as pd
import numpy as np
```

```python
from sklearn.feature_extraction import FeatureHasher
```

```python
d = {'Outlook': ['Humid', 'Sunny','Overcast','Rain','Rain','Humid','Overcast','Sunny','Sunny','Rain','Sunny','Humid','Overcast','Rain','Windy','Sunny','Overcast','Overcast','Rain','Windy'
,'Windy','Windy','Sunny','Rain','Humid','Rain','Overcast','Sunny','Windy','Windy','Humid','Humid','Humid']
}
df = pd.DataFrame(data=d)
```

```python
df['Outlook'].unique()
```

```python
fh = FeatureHasher(n_features=4, input_type='string')
hashed_features = fh.fit_transform(df['Outlook'])
hashed_features = hashed_features.toarray()
## Concateing the dataframes
pd.concat([df[['Outlook']], pd.DataFrame(hashed_features)], 
          axis=1).head()
```

#### Feature Hashing on Video Games Data

```python
df=pd.read_csv('vgsales.csv')
df.head(10)
```


Only taking Few elements for demonstrate purposes

Predictor Variables are 'Platform','NA_Sales','EU_Sales','JP_Sales','Other_Sales'

Target Variabe is Global_Sales


```python
df_x=df[['Platform','NA_Sales','EU_Sales','JP_Sales','Other_Sales']]
df_y=df[['Global_Sales']]
```

```python
## Feature Encoding
fh = FeatureHasher(n_features=10, input_type='string')
hashed_features = fh.fit_transform(df['Platform'])
hashed_features = hashed_features.toarray()
## Concateing the dataframes
df_X=pd.concat([df_x, pd.DataFrame(hashed_features)], 
          axis=1)
df_X.head()
```

```python
print('Droping categorical column platform')
df_X=df_X.drop(['Platform'], axis=1)
df_X.head()
```

#### Train-Test Split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df_X, df_y, test_size=0.3, random_state=422,shuffle=True)
print("Number Data in X_train dataset: ", X_train.shape)
print("Number Data in y_train dataset: ", y_train.shape)
print("Number Data in X_test dataset: ", X_test.shape)
print("Number Data in y_test dataset: ", y_test.shape)
```

#### Random Forest

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(n_estimators = 100, random_state = 42)
rf.fit(X_train, y_train);

# Use the forest's predict method on the test data
y_pred = rf.predict(X_test)
y_pred=(y_pred).reshape(4980, 1)
```

```python
from sklearn import metrics

print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))  
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))  
```

```python
errors = abs(y_pred - y_test)
# Calculate mean absolute percentage error (MAPE)
mape = 100 * (errors / y_test)
# Calculate and display accuracy
accuracy = 100 - np.mean(mape)
print('Accuracy with Hashing Trick:', round(accuracy[0], 2), '%.')
```
